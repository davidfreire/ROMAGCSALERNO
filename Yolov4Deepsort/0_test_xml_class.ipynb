```python
def print_frame(frame, patch):
    for idx in range(0, len(patch[frame]['instances']['images'])):
        img = patch[frame]['instances']['images'][idx]
        rp = patch[frame]['instances']['relative_position'][idx]
        
        print('--------------')
        print('Relative position in scene:', rp)
        height, width, depth = img.shape
        dpi = plt.rcParams['figure.dpi']
        figsize = width / float(dpi), height / float(dpi)
        plt.figure(figsize=figsize)

        plt.imshow(img)
        plt.title('Picture')
        plt.show()

```

```python
print_frame(0, patch) 
# Nothing detected on frame 0, {'Frame_ID': 1, 'Instances': []} 
```
